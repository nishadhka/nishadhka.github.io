## Trail for WRF-CHEM, WRF simulation with Katrina case and one way Nesting-Ndown.exe
- It is based on [this](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/CASES/NestRuns/ndown.htm) to run compiled wrf in serial mode to execute the Katrina case with nested model run by Ndown.exe based one way nesting.

### WPS: Ungrib

- As per this [page](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/CASES/SingleDomain/ungrib.htm), the WPS ungrib stage was carried out.
- The tar file provided with the tutorial was unzipped and compiled WPS folder was copied and both these folders were kept under a folder named katrina as katrina_nesting.
- The vtable was linked inside WPS folder by ```ln -sf ungrib/Variable_Tables/Vtable.GFS Vtable```.
- The avn/GFS files inside the tar folder is linked with WPS folder by ```./link_grib.csh /home/swl-sacon-dst/wrf/WRF341/katrina/GFS_data/```, by this following file with links were created by this step inside WPS folder.
```fortran    
GRIBFILE.AAA -> ../katrina/GFS_data/avn_050828_00_00 
GRIBFILE.AAB -> ../katrina/GFS_data/avn_050828_00_06  
GRIBFILE.AAC -> ../katrina/GFS_data/avn_050828_00_12  
GRIBFILE.AAD -> ../katrina/GFS_data/avn_050828_00_18  
GRIBFILE.AAE -> ../katrina/GFS_data/avn_050828_00_24  
GRIBFILE.AAF -> ../katrina/GFS_data/avn_050828_00_30  
GRIBFILE.AAG -> ../katrina/GFS_data/avn_050828_00_36  
GRIBFILE.AAH -> ../katrina/GFS_data/avn_050828_00_42  
GRIBFILE.AAI -> ../katrina/GFS_data/avn_050828_00_48  
```    
- For ungrib part of WPS a minimal namelist.wps suggested by this [page](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/Basics/UNGRIB/ungrib_namelist.htm) as part of the tutorial is used. By this the namelist.wps in WPS folder was edited and the file become
```fortran
&share
max_dom = 1,
start_date = '2005-08-28_00:00:00','2005-08-28_00:00:00',
end_date = '2005-08-30_00:00:00','2005-08-28_00:00:00',
interval_seconds = 21600,
io_form_geogrid = 2,
/
&metgrid
fg_name = 'FILE'
io_form_metgrid = 2,
/
```
- With this setup ungrib was run using this command ```./ungrib.exe >& ungrib_data.log```, the log file says successful completion of the ungrib. With the earlier version of file ungrib, given error of namlist read and namelist was edited then ungrib run without error.
- In tutorial the remaining steps with SST data, based on [this](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/DATA/SST/index.html) was not carried out. SST data is only required for long simulations.

```bash
sudo mount -t vboxsf -o uid=1000,gid=1000 share ~/host
```
