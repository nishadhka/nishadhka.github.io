## Trail for WRF-CHEM, working with WRF simulation with Katrina single domain case
1. It is based on [this](http://www.mmm.ucar.edu/wrf/OnLineTutorial/CASES/SingleDomain/index.html) to run compiled wrf in serial mode to execute the Katrina case with single domain.
2. As per this [page](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/CASES/SingleDomain/ungrib.htm), the WPS ungrib stage was carried out

### WPS: Ungrib
3. The tar file provided with the tutorial was unzipped and compiled WPS folder was copied and bothe these folders were kept under a folder named katrina.
4. The vtable was linked inside WPS folder by ```ln -sf ungrib/Variable_Tables/Vtable.GFS Vtable```.
5. The avn/GFS files inside the tar folder is linked with WPS folder by ```./link_grib.csh /home/swl-sacon-dst/wrf/WRF341/katrina/GFS_data/```, follwoing file with links were created by this step
```bash
GRIBFILE.AAA -> ../DATA/Katrina/avn_050828_00_00
GRIBFILE.AAB -> ../DATA/Katrina/avn_050828_00_06
GRIBFILE.AAC -> ../DATA/Katrina/avn_050828_00_12
GRIBFILE.AAD -> ../DATA/Katrina/avn_050828_00_18
GRIBFILE.AAE -> ../DATA/Katrina/avn_050828_00_24
GRIBFILE.AAF -> ../DATA/Katrina/avn_050828_00_30
GRIBFILE.AAG -> ../DATA/Katrina/avn_050828_00_36
GRIBFILE.AAH -> ../DATA/Katrina/avn_050828_00_42
GRIBFILE.AAI -> ../DATA/Katrina/avn_050828_00_48
```
6. For ungrib part of WPS a minimal namlist.wps suggestby this [page](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/Basics/UNGRIB/ungrib_namelist.htm) as part of the tutorial is used. By this the namelist.wps in WPS folder was edited as 
```bash
&share
start_date = '2005-08-28_00:00:00','2007-08-15_00:00:00',
end_date = '2005-08-30_00:00:00','2007-08-16_00:00:00',
interval_seconds = 21600
/
&ungrib
out_format = 'WPS',
prefix = FILE,
/
``` 
7. With this setup ungrib was run using this command `./ungrib.exe >& ungrib_data.log`, the log file says successful completion of the ungrib. In tutorial the remaining steps such as of SSt data, after point 7 is not carried out.

### WPS:Geogrid

8. This part of tutorial is based on the [page](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/CASES/SingleDomain/geogrid.htm)
9. For the geogrid part of WPS, following namelist.wps was used based on [this](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/Basics/GEOGRID/geogrid_namelist.htm) and namelist.wps becomes
```bash
&share
wrf_core = 'ARW',
max_dom = 1,
io_form_geogrid = 2,
/
&geogrid
parent_id = 1, 1,
parent_grid_ratio = 1, 3,
i_parent_start = 1, 31,
j_parent_start = 1, 17,
e_we = 74, 112,
e_sn = 61, 97,
geog_data_res = '10m','2m',
dx = 30000,
dy = 30000,
map_proj = 'lambert',
ref_lat = 34.83,
ref_lon = -81.03,
truelat1 = 30.0,
truelat2 = 60.0,
stand_lon = -98.0,
geog_data_path = '/home/hoopoe/ems/wrfems/data/geog/'
/
```
9. In namelist.wps, the line ```geog_data_path``` was edited with geog folder as per the EMS namlist.wps to refer the huge file required for this purpose.
10. There is a step stating the requirement of ncl language to visualize the domain, since no ncl language was installed it was omitted.
11. With this the geogrid was executed by ```./geogrid.exe``` and ended with successful completion as follows and a ```geo_em.d01.nc``` was created.
```bash
./geogrid.exe
Parsed 22 entries in GEOGRID.TBL
Processing domain 1 of 1
  Processing XLAT and XLONG
  Processing MAPFAC
  Processing F and E
  Processing ROTANG
  Processing LANDUSEF
  Calculating landmask from LANDUSEF ( WATER =  16 )
  Processing HGT_M
  Processing HGT_U
  Processing HGT_V
  Processing SOILTEMP
  Processing SOILCTOP
  Processing SCT_DOM
  Processing SOILCBOT
  Processing SCB_DOM
  Processing ALBEDO12M
  Processing GREENFRAC
  Processing SNOALB
  Processing SLOPECAT
  Processing SLOPECAT
  Processing CON
  Processing VAR
  Processing OA1
  Processing OA2
  Processing OA3
  Processing OA4
  Processing OL1
  Processing OL2
  Processing OL3
  Processing OL4
  Processing VAR_SSO
!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
!  Successful completion of geogrid.        !
!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
```

### WPS:Metgrid
 
12. This part of tutorial is based on the [page](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/CASES/SingleDomain/metgrid.htm)
13. For the metgrid part of WPS, following namelist.wps was used based on [this](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/Basics/METGRID/metgrid_namelist.htm) and namelist.wps becomes
```bash
&share
max_dom = 2,
start_date = '2005-08-28_00:00:00','2005-08-28_00:00:00',
end_date = '2005-08-30_00:00:00','2005-08-28_00:00:00',
interval_seconds = 21600,
io_form_geogrid = 2,
/
&metgrid
fg_name = 'FILE'
io_form_metgrid = 2,
/
```
14. The metgrid was executed by ```./metgrid.exe``` and it gives error 
```bash
Processing domain 2 of 2
 WRF_DEBUG: NetCDF error: No such file or directory
 WRF_DEBUG: NetCDF error in wrf_io.F90, line        1059
ERROR: Couldn't open file ./geo_em.d02.nc for input.
application called MPI_Abort(MPI_COMM_WORLD, 0) - process 0
```
it seems the namelist.wps is wrong by specifying ```max_dom =2``` so the namelist.wps was edited with ```max_dom =1```
15. By above edit metgrid.exe ended successful completion with creation of nc files
 ###WRF:real.exe and wrf.exe###
16. This part of tutorial is based on the [page](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/CASES/SingleDomain/wrf.htm)
17. For making WRF run the folder content from ```WRFV3_serial/test/em_real ``` was used as such and all the met files are linked inside the em_real folder using the command 
```bash
ln -sf /home/swl-sacon-dst/wrf/WRF341/katrina/kWPS/met_em.d01.2005* .
```
18. For the real.exe and wrf.exe part of WRF, following namelist.input was used based on [this]http://www2.mmm.ucar.edu/wrf/OnLineTutorial/Basics/WRF/namelist.input.htm) and edited to suit the katrina case and the namelsit.input becomes
```bash
&time_control
run_days = 0,
run_hours = 12,
run_minutes = 0,
run_seconds = 0,
start_year = 2005,
start_month = 08,
start_day = 28,
start_hour = 00,
start_minute = 00,
start_second = 00,
end_year = 2005,
end_month = 08,
end_day = 30,
end_hour = 00,
end_minute = 00,
end_second = 00,
interval_seconds = 21600
input_from_file = .true.,
history_interval = 180,
frames_per_outfile = 1000,
restart = .false.,
restart_interval = 720,
io_form_history = 2
io_form_restart = 2
io_form_input = 2
io_form_boundary = 2
debug_level = 0
/
&domains
time_step = 180,
max_dom = 1,
s_we = 1, 
e_we = 74,
s_sn = 1, 
e_sn = 61,
s_vert = 1,
e_vert = 30,
num_metgrid_levels = 27
dx = 30000, 
dy = 30000, 
grid_id = 1,
parent_id = 1,
i_parent_start = 1, 
j_parent_start = 1, 
parent_grid_ratio = 1,
parent_time_step_ratio = 1,
feedback = 1,
/
&physics
mp_physics = 3,
ra_lw_physics = 1, 
ra_sw_physics = 1, 
radt = 30, 
sf_sfclay_physics = 1, 
sf_surface_physics = 1,
num_soil_layers = 5,
bl_pbl_physics = 1, 
bldt = 0, 
cu_physics = 1, 
cudt = 5, 
/
&dynamics
w_damping = 0,
diff_opt = 1,
km_opt = 4,
diff_6th_opt = 0,
diff_6th_factor = 0.12,
base_temp = 290.
damp_opt = 0,
zdamp = 5000., 
dampcoef = 0.2,
khdif = 0, 
kvdif = 0, 
non_hydrostatic = .true., 
moist_adv_opt = 1,
scalar_adv_opt = 1,
/
&bdy_control
spec_bdy_width = 5,
spec_zone = 1,
relax_zone = 4,
specified = .true.,
nested = .false., 
/
```
19. With this setup the real.exe was executed by ```./real.exe ``` and error was bought about missing met file, then it is found that the above namelist content was wrongly mentioning start day date and so namelist.input was edited and real.exe ends with successful completion with creation of two files ```wrfinput_d01 and wrfbdy_d01``` each around 9.0 MB and 21 MB respectively.
20. Then real.exe was executed by ```./wrf.exe``` and it completed with out any error and successfully ends with creation of two files as per the tutorial 
```bash
wrfout_d01_2005-08-28_00:00:00
wrfrst_d01_2005-08-28_12:00:00
```
