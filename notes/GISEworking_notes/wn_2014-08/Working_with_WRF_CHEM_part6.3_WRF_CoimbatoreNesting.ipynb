## Trail for WRF-CHEM, WRF simulation with domain for Coimbatore and one way Nesting-Ndown.exe
- It is based on [this](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/CASES/NestRuns/ndown.htm) to run compiled wrf in serial mode to execute domain over Coimbatore with nested model run by Ndown.exe based one way nesting.

### WPS: Ungrib

- As per this [page](http://www2.mmm.ucar.edu/wrf/OnLineTutorial/CASES/SingleDomain/ungrib.htm), the WPS ungrib stage was carried out.
- The gfs files downloaded was used, average size around 45MB for each files. The files for time peroid 00,03, 06 was kept under a folder gfs.
- The vtable was linked inside WPS folder by ```ln -sf ungrib/Variable_Tables/Vtable.GFS Vtable```.
- The GFS files inside the gfs folder is linked with WPS folder by ```./link_grib.csh /home/swl-sacon-dst/wrf/WRF341/coimbatore/gfs/```, by this following file with links were created by this step inside WPS folder.
```fortran 
GRIBFILE.AAA -> /home/swl-sacon-dst/wrf/WRF341/coimbatore/gfs/14060500.gfs.t00z.pgrb2f00 
GRIBFILE.AAB -> /home/swl-sacon-dst/wrf/WRF341/coimbatore/gfs/14060500.gfs.t00z.pgrb2f03
GRIBFILE.AAC -> /home/swl-sacon-dst/wrf/WRF341/coimbatore/gfs/14060500.gfs.t00z.pgrb2f06
```     
-  The namelist.wps in WPS folder was edited and the file become
```fortran
&share
max_dom = 4,
start_date ='2014-06-05_00:00:00','2014-06-05_00:00:00','2014-06-05_00:00:00','2014-06-05_00:00:00'
end_date = '2014-06-05_06:00:00','2014-06-05_06:00:00','2014-06-05_06:00:00','2014-06-05_06:00:00'
interval_seconds = 10800,
io_form_geogrid = 2,
/
&metgrid
fg_name = 'FILE'
io_form_metgrid = 2,
/
```

-  The ungrib.exe was run using this command ```./ungrib.exe >& ungrib_data.log```, the log file says successful completion of the ungrib.

- It is found in wrf.exe in former case where wrfoutput was returned in residual timings such as this ```Timing for Writing wrfout_d01_2014-06-05_05:56:24 for domain        1:    0.50122 elapsed seconds```, the error is suspected to realted with this writing. So amde a serach to output the wrf in every hour rather than sigle file as produced in former wrf.exe for coimbatore nesting. Foun this [page](http://www.dtcenter.org/wrf-nmm/users/OnLinePractical/boston_nest.htm) indiating the naelist input variable to be changed into make wrf output in every hour. As per the page the variables under &time_control such as history_interval = 60, 60, [Outputs every hour] and frames_per_output = 1, 1, [Outputs one file per hour - desired for UPP] was edited in the namelist.input. surprisingly the varibles for frames_per_output was arround 1000 in the former running namelist.input. So the namelist.input was edited and the file is [this](namelist.input_DO1_v2). The created files in later run was deleted and rerun the real.exe and wrf.exe for domain DO1.
-  The editing of namelist.input variabel such as history_interval and frames_per_output are making wrf to output in per hour but it didn't changed the problem of rounding the time in nearest hour ending, the problem faced by ndow.exe. Searched with this and found no information regarding this error and followed the referred page namelist in which the time step is rounded to 180 seconds that way it produces wrf out put in rounded hours, other way as per the wrf ems namelist.real and the pref‏ered method for choosing time step with 6x dx where dx is grid resolution in km is not able to follow, if it is given for 27 km resolution with time step of 6x27=162, it create wrf output in residual timings for example, first hour in wrfout*01:02:21. 
-  Using the formed wrf ouput in rounded hour, it is moved further to nesting for second domain DO2, followed the ```making temporary fine grid initial condition- Domain 02```, renamed the met_em* files and edited the namelist.input for domain DO2 [the file](namelist.input_DO2_v2). Run ./real.exe and it created the files ```wrfinput_d01, wrfbdy_d01```, in which ```
