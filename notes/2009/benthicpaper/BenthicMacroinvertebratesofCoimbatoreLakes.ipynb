**Abstract:** Urbanization has profound and long-lasting impact on wetlands, leading to the total deterioration of the whole ecosystem. This complex and dynamic problem needs in depth understanding to foster conservation and management measures. We investigated the effect of urbanization along a typical rural to urban gradient wetlands of Coimbatore, using physico chemical, biological assessment and spatial analysis methods to deduce current status and spatial analysis to infer pattern of urbanization impacts. Physico-chemical analyses of wetland water revealed increased degradation in water quality in those wetlands which were in close proximity to urban centers. The littoral benthic fauna communities of these wetlands were similar between the wetlands but showed significant variation in community composition and abundance along the urbanization gradient. Land cover change and spatial analysis indicate that there was an unprecedented expansion in urban area around these wetlands and it was higher around more degraded wetland. Hence urbanization can be linked to the current status of these wetlands. The results indicate the negative effect of urbanization on wetlands.. 

**Keywords:** Benthic macro invertebrates, Coimbatore, Landsat, Urbanization Impact, Wetlands.

## Introduction:
In the scenario of ongoing rapid urbanization, especially in developing countries, its role in depriving the  ecosystem goods and services has to be better understood for devising and implementing appropriate environmental management options (Alberti, 2005). The process of urbanization starts from conversion of rural land into suburban, urban or  industrial land use through change in land occupancy and use. This process generally leads to increase in human population, its establishment densities and deleterious impact on ecosystem within its periphery (Lindh, 1972). Sustaining a diverse biota through provision of habitat and influencing local climate and hydrology make wetlands among the most productive ecosystems on earth (Whittaker,1971, Adams and Stockwell, 1983). Wetlands in the vicinity of expanding urban conglomerate pose immense threat through pollution, encroachment and catchments alteration (Ramachandra, 2008 and Du et al, 2010). Other than  a major cause of biodiversity loss, urbanization impacts aquatic ecosystems through degrading and disrupting the energy, nutrient, and hydrological cycles, and homogenizing biotic components in it (McKinney, 2002 and Alberti, 2005). 
Land use/land cover change from rural to urban has deleterious effect on wetlands through increasing point/non point pollution sources and its intensity of impact (Tsegaye et al, 2005). It became prominent effect on wetlands by deteriorating water quality mainly to nutrient enrichment, oxygen reduction, heavy metal contamination and increased siltation (Mohanraj et al, 2000). Temporal accumulation of those pollutants in water column and sediment of wetland culminate into a situation of unsuitable habitat for aquatic and dependent organism resulting in collapse of ecological networks for sustaining the ecosystem (Booth and Jackson, 1997). Benthic macroinvertebrate  is a  crucial connecter between producers and carnivores of wetland ecosystem and their community composition reflects the ecosystem health and indicates  influential factors especially water quality affecting their composition (Cooper et al., 2007).  Their prevalence  is dependent on habitat conditions which are determined by water quality, habitat quality, energy source, flow regime and other biotic interaction factors (Wiederholm, 1980). Assessment of these aquatic organisms along with water quality of the wetland may help in revealing the systems’ current functionality and its status (Spellerberg, 2005).  Wetlands, with their characteristic features of integrating inputs from diverse sources, nonlinear and dynamic response, and slow recovery from external drivers of change, understanding their threats from urbanization and devising management will not be comprehensive without inclusive basin level approach (IECC,2005). Land cover change study using multi-temporal satellite imagery is a promising technique for basin level urbanization change analysis  (Jensen, 2000). It gives an extensive opportunity in assessing the temporal changes in land use/cover pattern in the vicinity of wetland, an important factor determining the health of the system. Further spatial analysis using land use land cover change study gives insight into how those geomorphological changes might affect the wellbeing of wetland as an ecosystem. Thus it enhances the basin level approach of understanding the impact of urbanization on wetlands.

Coimbatore, the eleventh largest city in the country and among the fastest growing one is blessed with river Noyyal, formerly a perennial one feeding around 28 wetlands within the urban sprawl. The growing industrialization and ever expanding urbanization pose considerable strain on basic amenities of the city. The city wanting facilities for treatment of industrial and municipal waste water has open drainage and sewerage systems connected to nearby streams and lakes without any prior treatment. Earlier studies carried out on these wetlands document the ecological effects of the point source pollution. These studies have also documented issues related with urban life such as disposal of solid waste in wetlands, removal of vegetation, defecation practices, and encroachment (Mohanraj et al, 2000, Mathew et al, 2002 and Chandra et al, 2010). In view of these previous studies, the present one was carried out to assess the prevailing response profile of four typical rural to urban gradient wetlands towards urbanization using physiochemical and benthic macro invertebrate bio-assessment methods. The urbanization analysis comprised of land use, land cover change assessment and spatial analysis was used to deduce causes of apparent pattern of wetland response towards urbanization.

## Materials and Methods:
### Sampling sites:
Three sampling sites were selected in each of the four wetlands namely Perur, Kurchi, Ukkadam, and Sengulam (Figure-01). Sampling was conducted in February 2009 (See Chandra et al, 2010 for detailed description of lakes).
### Benthic Macro invertebrate and water sampling:
Samples for benthic macro-invertebrate analysis were collected in triplicates from littoral zones by using Van Veen’s Grab sampler with surface area of 0.0318m2. Immediately after collection sediments from the sampler was transferred into 20 liter plastic bucket and the sampler was rinsed with filtered wetland water into the same bucket. The sediment sample was made into suspension with water and filtered through series of mesh, from coarse to finest of mesh size 0.5mm. The filtration was continued until no further organism in the sediment sample was found. The organisms adhered to the mesh were collected and preserved in 4% Formalin solution. In laboratory, the specimens were counted and identified using keys of Mellanby (1938) and Edmondson (1959) then cross referred with the Integrated Taxonomic Information system (ITIS). Identification was done up to family level wherever possible, but only up too order for Plecoptera and Ostrocoda. Eleven physico chemical parameters of wetland water were measured at each benthic sampling site. Samples for these analyses were collected in pre-cleaned bottles, labeled, and immediately transferred to in ice boxes. The samples in ice boxes were transported to the laboratory, stored at 4°C in dark till further processing and analysis. Physico-chemical parameters namely pH, electrical conductivity (EC, µS/cm), total dissolved solids (TDS, mg/L), total hardness(mg/l), biological Oxygen demand (BOD, mg/L), alkalinity (mg/L), chloride (mg/L), sulphate (mg/L), phosphate (mg/l), ammonia (mg/l), sodium (mg/l), Potassium (mg/l) were examined following standard methods (Greenberg et.al 1998).

### Land use classification and spatial analysis:
The satellite imagery for land cover classification by NASA’s Landsat TM for 1989 and Landsat ETM+ for 1999 was obtained from the Global Land Cover Facility (GLCF –www.landcover.org). The geographical area of Coimbatore was included in the world reference system 2 path 144 and row 052 of Landsat imagery. Processing and image analysis was carried out using the free open source Geographical Information System (GIS) software GRASS (GRASS Development Team, 2008). The land cover was classified by supervised classification algorithm using Maximum Likelihood classifier. Land cover / land use summery grouped into urban and nonurban land cover was tabulated (Neteler and Mitasova, 2007). Spatial analysis was carried out using Spatial Analyst Extension of Arc GIS 9.2 software. The classified image of study area was vectorised as Polygon feature for wetland boundary and point features for urban area using raster to vector tool in the spatial analyst tool bar. The point features of urban area were subjected for density calculation and reclassification options in spatial analyst toolbar. The highest density urban area was vectorized and Euclidian distance analysis was done among the urban centers and wetland boundary. The map generated from Euclidian distance analysis and layer of vectorised wetland boundary was stacked for generating urban proximity map which indicate the degradation potential on wetlands from urban centers.
Statistical Analysis:
Benthic and physico chemical characteristics of water samples data was log normalized and used for Principle Component Analysis (PCA) and the results were  represented in biplot ordination diagram. PCA was used for visualizing gradient of benthic community and water quality in the four lakes (Shaw, 2003). Pearson correlations were calculated between first principle component scores and physico-chemical characteristics of water samples to determine the variables that were weighted heaviest in the principle component. One tailed ANOVA was carried out to assess the variability in PC scores. All the statistical analysis was carried out in open source statistical package R using community ecology package vegan (Dixon, 2003) 
## Results and Discussion:
### Benthic macroinvertebrate and water quality:

The PCA on physico-chemical characteristics of water samples (Table-1) shows that the first two principle components (PCs) explained 76.3% of the variability in data and the biplot visualized clear gradation in water characteristics between suburban  and urban wetlands (F3,20=4.81, P=0.011, Figure-02). The PC1 of PCA represented the extent of water pollution in which negative scores indicates high value for certain parameters and thus high pollution and positive scores indicates low pollution. The vector length of water characteristic in biplot indicates the influence of respective water parameter on scores evident from significant negative correlation between PC1 scores and the water characteristics (p<0.01) except BOD(r=-0.433, P=0.16), ammonia(r=-0.389,P=0.211) and  positive correlation with Phosphate (r=0.3,P=0.343). The biplot shows clear clumping between Chinnakulam, Perur wetlands and Ukkadam wetland . Among these, former wetlands recorded lowest value for all parameters except phosphate while later one recorded vice versa, showing the extent of degradation in terms of water quality in these wetlands (Table-1). The wetland of Kurchi stands in between the range and it responded site specifically to water quality impairment in which the site1 was relatively more polluted than other two sites.

We collected benthic macro invertebrates representing 24 taxa from these wetlands (Table-2). Highest number of taxa was recorded from Kurchi (17 taxa) followed by Perur (14 taxa), Chinnakulam (14 taxa) and Ukkadam (13 taxa, Table-2) Chironomidae (Diptera), the most abundant taxon in all the four wetlands, highest percentage abundance was observed in Ukkadam (68%) followed by Kurchi (32%), Chinnakulam (30%) and Perur (29%). In Perur the next most abundant taxa are Tricorythidae (23%) and Corixidae (10%). In Kurchi and Ukkadam it was by Lymnaeidae (27% and 13% respectively). In the case of Chinnakulam it was by Tricorythidae (14%) and Lymnaeidae (10%, Table-2). The influence of urbanization on benthic macro invertebrates was reflected in the dominance of pollution tolerant groups such as  Chironomidae and Lyminidea in urbanized Ukkadam and Kurchi wetlands. In less urbanized Perur and Chinnakulam lakes the pollution tolerant benthoses were lesser and pollution sensitive groups such as Tricorythidae and Corixidae (Mandaville, 2002, Sharma, 2008, Stanley and James, 1997) constituted the benthic fauna largely. Principle component analysis of benthic faunal data of these four wetlands shows that the gradation was observed in between positive PC2 to positive PC1 scores (Figure-3).; the two sites of Perur and all sites of Kurchi, Ukkadam clumped together to form two distinct clusters. The remaining sites of Kurchi and all sites of Chinnakulam fall in between these clusters.

The high value of all the water parameters in Ukkadam, compared with rest of the wetlands show high load and long restraint of pollution from urban centers. Industrial effluents such as dyeing, bleaching and sewage effluents draining into Ukkadam contain high quantity of salts, nutrient and heavy metals (Mohanraj et al, 2000, Senthilnathan and Azeez, 1999). The benthic zone nearer to the effluent inlet of Ukkadam was completely devoid of benthic community, a situation that can be attributed to persistent accumulation of pollutants and their secondary products in the wetland (Mathew et al, 2002, 2003). Though the Kurchi lake was surrounded by urban sprawl its pollution profile was considerably lower than Ukkadam and it sustains most rich benthic communities among all the four wetlands. The relative shallowness, wide variety of flora, lesser municipal solid waste and industrial effluent and earthen embarked detention pond for storing sewage effluent are determining the better profile of Kurchi lake though it is a lake surrounded by urban land use. In suburban Perur and Chinnakulam wetlands phosphate concentrations were relatively higher especially during summer seasons (Chandra et al, 2010). It can be specifically attributed to agriculture and consequent use of Diammonium Phosphate (DAP) in the adjoining areas of these wetlands.  The benthic community composition in these wetlands was rather similar reflecting the a better stateg compared to other urbanized wetlands.   
  

### Urbanization analysis:

To assess the urbanization through land use and land cover change, an area of 30 Sq.km was selected that covered the wetland area and urban centers surrounding those wetlands. The false color composite image of study area along with classified image of land use for study area during 1989 and 1999 is shown in Figure 4. It was observed that in the study area in 1989 urban area was about 486.4 hectares and in 1999 it was about 735.9 hectares (Table 3). The non-urban area comprising water body, vegetation and bare land was about 3373.8 hectares in 1989 and 3124.2 hectares in 1999. Thus, in a decade the urban area was increased by about 51.2% in the study area (Table 3). In the same span (1991-2001), urban population in Coimbatore district increased by 51.1% while urban density increased by 14% (Balasubramanian and Choi, 2010).  The urban proximity map generated (Figure 4E) by spatial analysis of 1999 classified imagery shows that Ukkadam has more number of urban pockets in its immediate vicinity and is likely to be highly impacted by urban degradation. In the case of Kurchi the urban pockets were relatively lesser than that of Ukkadam. The urban pockets were relatively far in the case of other two lakes currently showing low degradation levels.

Coimbatore  a lake centric city, the wetlands in it, largely developed by the erstwhile rulers, were the main drivers for originating of its urban pockets. In later years during the development of the city Ukkadam remained the most accessible point for people from different areas as it lies more or less in the opening of the Palghat gap, in the Western Ghats, to the city. Thus it became proximate to markets and transports. It is also a city hub accessible to most dense population centers of Coimbatore. Likewise Kurchi is surrounded by municipalities Kurchi and Kuniamuthur. Since the urban pockets surrounding the Ukkadam wetland are the center of urban expansion, it is profoundly and historically impacted resulting in its current status. Though the Perur wetland has urban pocket in its vicinity well before Ukkadam wetland, being a historic temple town agriculture based economy have prevailed and the wetland is considered as an important water source for its economy. Perur has restrained itself from becoming industrialized for its traditions. On the other hand Ukkadam has grown up to be a transport hub and industrialized economy mainly dealing with textile industry of dyeing and bleaching where the  wetland is considered only as a disposal site industrial discharges. This negligence of other ecological services from the wetland is detrimental.
