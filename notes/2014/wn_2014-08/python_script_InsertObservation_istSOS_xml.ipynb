## python script to InsertObservation into istSOS_xml way

1. Based on the early experience of inserting observation in xml way with python using HTTP post in 52North SOS, made a try with Ist SOS, other way to do is using calling another python script to import data in csv formate. 
2. Based on this note on OWS service [requests](https://geoservice.ist.supsi.ch/projects/istsos/index.php/Example_of_SOS_request). to get the DescribeSensor information used this link, it always failed in specfying formate but other request was succeffull
```
http://54.255.173.125/istsos/cbed?request=describeSensor&procedure=KNMR&outputFormat= text xml;subtype="sensorML/1.0.1"&service=SOS&version=1.0.0
```
1. to GetFeatureInterest
```
http://54.255.173.125/istsos/cbed?request=getFeatureOfInterest&featureOfInterestId=KNMR&srsName=4326&service=SOS&version=1.0.0
```
1. using the lins http://localhost/istsos/demo/?request=getFeatureOfInterest&featureOfInterestId=LUGANO&srsName=21781&service=SOS&version=1.0.0 and http://localhost/istsos/demo?request=describeSensor&procedure=T_LUGANO&outputFormat=text/xml;subtype=%27sensorML/1.0.0%27&service=SOS&version=1.0.0
1. to GetObservation
```
http://54.255.173.125/istsos/wa/istsos/services/cbed/operations/getobservation/offerings/temporary/procedures/SVNC/observedproperties/urn:ogc:def:parameter:x-istsos:1.0:cbe:dylos:np05mg,urn:ogc:def:parameter:x-istsos:1.0:cbe:dylos:np25mg/eventtime/2014-04-29T00:00:00+0530/2014-04-30T00:00:00+0530?_=1405529585826
```
3. From GetObservation request found the basic detials requered for InsertObservation request they are
```
procedure: "urn:ogc:def:procedure:x-istsos:1.0:KNMR", 
featureOfInterest:"urn:ogc:def:feature:x-istsos:1.0:Point:KNMR", observedProperty":    "urn:ogc:def:parameter:x-istsos:1.0:time:iso8601",
"urn:ogc:def:parameter:x-istsos:1.0:cbe:dylos:np05mg",
"urn:ogc:def:parameter:x-istsos:1.0:cbe:dylos:np25mg",
```
4. So using above details made xml for InsertObservation, it is became
