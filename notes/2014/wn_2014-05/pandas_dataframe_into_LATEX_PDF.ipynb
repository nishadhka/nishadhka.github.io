**pandas_dataframe_into_PDF**
- Huge dataframes whihc goes for multiple A4 pages landscape is difficult to make in excel
- Alternative is use python pandas and convert the pandas dataframe into pdf thorugh latex or html, latex is promising for just printing
- Basically from [1](http://stackoverflow.com/questions/14380371/export-a-latex-table-from-pandas-dataframe) for convert dataframe into tex and   [this](http://tex.stackexchange.com/questions/147504/how-to-make-a-latex-document-in-landscape) for converting tex into landscape pdf document
- the python script to make table text is as follows, It is mostly from 3rd answer [1] and bold column heading write python trick from [3](http://techoverflow.net/blog/2013/12/08/converting-pandas-dataframe-to-latex-tabular/) 
```python
import pandas as pd
data = pd.read_csv('PROJECT_EXPNDTR1.csv')
data2 = data[data['Consumbales'] == 1]
with open("my_table.tex", "w") as f:
    f.write("\\begin{tabular}{" + " | ".join(["l"] * len(data2.columns)) + "}\n$
    columnLabels = ["\\textbf{%s}" % label for label in data2.columns]
    f.write ("%s\\\\\\hline\n" % " & ".join(columnLabels))
    \\writes column labels
    for i, row in data2.iterrows():
        f.write(" & ".join([str(x) for x in row.values]) + " \\\\\n")
    f.write("\\end{tabular}")
    f.close()
```
- this script makes table tex file and it can be opened in texstudio to add with oter latex comand to make a A4 landscape pdf
- the landscape A4 page is generated by latex command from [2] and added with tricks of longtable, columns size specification and nice all page heading row with ```\endhead``` based on ([4])(http://tex.stackexchange.com/questions/11380/how-to-repeat-top-rows-column-headings-on-every-page), latex goes like this
```bash
\documentclass{article}
\usepackage[a4paper,margin=1in,landscape]{geometry}
\usepackage{longtable}
\usepackage{tabulary}
\usepackage{blindtext}
\begin{document}
%% for adjust columns size\begin{longtable}{l | c | p{5cm} | p{2cm}| p{2cm} |p{2cm}| p{2cm}| p{2cm} |p{2cm} | p{2cm} | p{2cm}}%%
%%%python gernated tex file content column heading%%
\endhead
%%%python gernated tex file content rows%%
\end{longtable}
\end{document}
```
- makes a A4 landscape table outof dataframe
